列表S结构：

一个指针x，指向一条数据（理解成x就是这个数据）

一条数据包含一个key和从属数据

整个表包含n条记录

目标：

insert(S,x),向x数据插入内容

delete(S,x),删除x指向的数据

search(S,k),x对应的key是k的时候返回对应数据值

直接插入表

使用条件：key数量足够小

假设从含有m个元素的集合U中提取关键字（关键字唯一）

设定一个m元数组T来替代动态集合S，使得

如果x在S中，且x对应的key为k，那么T[k]=x,否则返回null

比如说有一条记录x=15，关键字是k，那么如果x存在于S中，T[k}返回15，否则返回null

耗时是O(1)

问题是如果S其实只有几千条数据，但是key是很大的数字，或者字符串，那么要维护的数组T非常庞大，浪费空间

散列函数：H 部分随机地映射向某些关键字到表T的下标（slot）

问题：不同的两个key被映射到同一个下标（冲突）

链接法解决冲突问题：

将同一个下标的记录，链接为一个链表

比如关键字49，86，52，都被映射到i下标，那么带入哈希函数后

h(49)=h(86)=h(52)=i

最坏情况，所有元素都被映射到同一个下标，变成一个巨大的链表

插入、删除、搜索时间：O(n)

平均情况：简单一致散列：假设所有关键字随机映射到各个下标

关键字有n个，下标有m个的散列表，装载因子为n/m

搜索期望时间为O(1)，条件是n=O(m),m跟得上n的时候



快速哈希：

h(k)=k mod m，用key对下标数m取模

缺点，假设正好key都是偶数，m也是偶数，所有数据都被散列到偶数下标

假设二进制数key，m正好是2的r次方，那么散列下标只有key的后r位决定

所以m通常希望选择是质数，不要接近2和10的幂

乘法哈希：

假设取m=2的r次方，计算机是w位的

h(k)=(A * k mod 2^w) rsh (w-r) 

A是w-1到2的w次方之间的奇数

rsh：右移位数（按位右移操作）

A不要太靠近w-1和2的w次方

例子：m=8，r=3，w=7

A=1011001

k=1101011

A * k = 10010100110011

对2^w取模：忽略高阶位，得到0110011

右移w-r=4位，得到剩下的3位011，就是下标

任何一个哈希函数都能找到一个情况让其运算最糟糕



解决冲突的方法：链表&开放地址法

开放地址法：

无需储存链表链接

如果哈希后的值不是空的，那么就用另一个哈希函数的值对应的位置，直到找到位置，索引的时候按照同样的顺序使用哈希函数即可

另一个哈希函数：只需要原来的哈希函数，增加一个参数表示探查次数

即哈希函数把key和探查次数的叉乘映射到散列表位置空间

需要保证元素数量小于等于表数量

缺点：难以处理删除，如果删除了，别人查找的时候，发现是空位置，会以为key不在这里，继续搜索（歧义）

参数i的设计：

H(k,i)=(h(k,0)+i)mod m

会有主簇现象，聚集区搜索很慢

双重散列：

H(k,i)=(H_1(k)+i * H_2(k))mod m

用两个散列函数组合

强制H_2为奇数，m为2的幂

评价搜索需要O(1)，随着表越满，速度会越慢，所以一般不能满占用


散列表的问题：对一个散列函数来说一定存在一个key集合能让其退化为链表

解决：随机散列函数（全域散列）


全域散列函数集H：设H为散列函数集合，m为散列表大小，对于不同的两个key，随机选择散列函数时，两个key碰撞的概率为1/m

假设我们构造出了全域散列函数集H，如果我们要把n个key映射到m个槽位上，对于一个key，与其发生碰撞的key的数量的期望小于n/m，装载因子，即key可以均匀分布

新问题：如何构造全域散列函数集

构造方法1：

思路：假设m为质数，用key mod m，依次取得其最低位保留为k_i，然后去掉最低位，用剩下的部分继续计算k_i+1

1.对key以m为基，取得其各位上的数字k0到kr，组成向量k

2.从0到m-1中随机选择a0到ar，组成向量a

3.计算散列函数集合h_a(k)=(sum(a_i * h_i)) mod m

相当于将a和k点乘再对m取模，每随机生一个a，就有一个对应的散列函数h_a

枚举所有a的可能，h_a组成的集合H总共有m^(r+1)个函数



构造方法2：完全散列

全域：关键字不确定，关注平均性能

完全：关键字确定，不插入不删除，关注最差性能，静态哈希表

要求在最坏情况下搜索O(1)

思路：构建二级散列，2级中都使用全域散列，用二级代替链表，每一个一级散列表的槽都对应一个二级散列表，最坏情况搜索O(1)
